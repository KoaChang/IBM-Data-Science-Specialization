```python
# this project analyzes the McDonalds menu nutritional info to make some cool insights
```

```python
import pandas as pd

df = pd.read_csv('menu.csv')
df.head()
```

```python
def calc_ratio(row):
    calories = row['Calories']
    protein = row['Protein']
    
    if calories == 0 or protein == 0:
        return 0
    else:
        return protein/calories
```

```python
df['Protein_To_Calorie_Ratio'] = df.apply(calc_ratio,axis=1)
df.head()
```

```python
top10 = df.nlargest(10,'Protein_To_Calorie_Ratio')
```

```python
top10_items = top10['Item']
top10_items
```
